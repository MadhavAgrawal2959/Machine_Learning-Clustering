```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
X=dataset.iloc[:,[3,4]].values
print(X)
```

##Using the DendoGram meathod to find the no of Cluster to be Made


```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("dendogram")
plt.xlabel("Cusomers")
plt.ylabel("Euclidean distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

#from the above dendrogram we can see that we can either use 3  cluster and 5 cluster the length was same

```python
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=60,c='red',label='Cluster 1')#here s is size of point while cis the color of the point of cluster
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=60,c='green',label='Cluster 2')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=60,c='blue',label='Cluster 3')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=60,c='brown',label='Cluster 4')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],s=60,c='pink',label='Cluster 5')
plt.title("(HierarchialClustering)Cluster of Customer")
plt.xlabel("Annual Income")
plt.ylabel("Spending score (1-100)")
plt.legend()
plt.show()
```
