```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
X=dataset.iloc[:,[3,4]].values
```

##Using the elbow meathod to find the optimal number of cluser


```python
from sklearn.cluster import KMeans
wcss=[];
for i in range(1,11):
  kmeans=KMeans(n_clusters = i, init='k-means++',random_state=42) ## we use k-means++ so that we don't fall into random initialization trap
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("the elbow meathod")
plt.xlabel("numbers of clusters")
plt.ylabel("wcss")
plt.show()
```

##Training the K-means model on the dataset


```python
  kmeans=KMeans(n_clusters = 5, init='k-means++',random_state=42) ## we use k-means++ so that we don't fall into random initialization trap
  y_kmeans=kmeans.fit_predict(X)
  print(y_kmeans)
```
